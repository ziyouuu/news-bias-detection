```python
import pandas as pd
df = pd.read_csv('../datasets/news_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['image', 'nlp-image_label'])
df.head()
```

```python
#lowercase
df['article_text'] = df['article_text'].apply(lambda x: x.lower())
```

```python
# remove punctuation
import re

def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df['article_text'] = df['article_text'].apply(remove_punctuation)
```

```python
#encoding for label 

df = pd.get_dummies(df, columns=['nlp_label'], dtype=int)
df.head()
```

```python
#tokenise
import nltk
from nltk.tokenize import word_tokenize

df['article_text_tokens'] = df['article_text'].apply(word_tokenize)
```

```python
# stopwords
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
article_text_tokens_no_stopwords = []
for row in df['article_text_tokens']:
    filtered_row = []
    for w in row:
        if w not in stop_words:
            filtered_row.append(w)
    article_text_tokens_no_stopwords.append(filtered_row)

df['article_text_tokens_no_stopwords'] = article_text_tokens_no_stopwords
```

```python
# bigram trigram 
from nltk.util import ngrams

def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))

# Generate n-grams for n = 2 (bigrams) and n = 3 (trigrams)
df['article_text_bigrams'] = df['article_text_tokens_no_stopwords'].apply(lambda x: generate_ngrams(x, 2))
df['article_text_trigrams'] = df['article_text_tokens_no_stopwords'].apply(lambda x: generate_ngrams(x, 3))
```

```python
df.head()
```

```python
sentiments_df = pd.read_csv('../datasets/generated/news_dataset_sentiment_labels.csv')
framing_df = pd.read_csv('../datasets/generated/classified_articles_with_political_bias.csv')
topic_df = pd.read_csv('../datasets/generated/articles_with_dominant_topic.csv')
```

```python
# Merge the DataFrames on the 'unique_id' column
merged_df = pd.merge(df, sentiments_df[['unique_id', 'negative', 'neutral', 'positive']], on='unique_id', how='left')
merged_df = pd.merge(merged_df, framing_df[['unique_id', 'frames', 'entities']], on='unique_id', how='left')
merged_df = pd.merge(merged_df, topic_df[['unique_id', 'dominant_topic']], on='unique_id', how='left')
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df_cleaned = merged_df.dropna()
```

```python
merged_df_cleaned.info()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from scipy.sparse import hstack
from scipy.sparse import csr_matrix

# Function to join lists into strings for vectorization
def join_lists_to_string(list_column):
    if isinstance(list_column[0], list):
        # list of lists
        return [' '.join(['_'.join(tup) if isinstance(tup, tuple) else str(tup) for tup in lst]) for lst in list_column]
    else:
        # single list
        return ' '.join(['_'.join(tup) if isinstance(tup, tuple) else str(tup) for tup in list_column])


# Convert list columns to strings
merged_df_cleaned['article_text_tokens_no_stopwords_str'] = join_lists_to_string(merged_df_cleaned['article_text_tokens_no_stopwords'])
merged_df_cleaned['article_text_bigrams_str'] = join_lists_to_string(merged_df_cleaned['article_text_bigrams'])
merged_df_cleaned['article_text_trigrams_str'] = join_lists_to_string(merged_df_cleaned['article_text_trigrams'])



```

```python
import ast

def concatenate_frames(frames_column):
    """
    Converts stringified lists into actual lists, then concatenates elements into a single space-separated string.
    """
    return frames_column.apply(lambda x: ' '.join(ast.literal_eval(x)) if pd.notna(x) else '')

merged_df_cleaned['article_text_frames_str'] = concatenate_frames(merged_df_cleaned['frames'])
merged_df_cleaned['article_text_entities_str'] = concatenate_frames(merged_df_cleaned['entities'])

```

```python
# Initialize vectorizers
vectorizer_tokens = TfidfVectorizer()
vectorizer_bigrams = TfidfVectorizer()
vectorizer_trigrams = TfidfVectorizer()
vectorizer_frames = TfidfVectorizer()
vectorizer_entities = TfidfVectorizer()


# vectorized features
tokens_features = vectorizer_tokens.fit_transform(merged_df_cleaned['article_text_tokens_no_stopwords_str'])
bigrams_features = vectorizer_bigrams.fit_transform(merged_df_cleaned['article_text_bigrams_str'])
trigrams_features = vectorizer_trigrams.fit_transform(merged_df_cleaned['article_text_trigrams_str'])
frames_features = vectorizer_frames.fit_transform(merged_df_cleaned['article_text_frames_str'])
entities_features = vectorizer_entities.fit_transform(merged_df_cleaned['article_text_entities_str'])
```

```python
# trigram 20000 features
```

```python
# Convert only sentiment features to sparse matrix
sentiment_features = csr_matrix(merged_df_cleaned[['negative', 'neutral', 'positive']].values)

# Handle topic if it's a numeric column
topic_features = csr_matrix(merged_df_cleaned[['dominant_topic']].values)

final_features = hstack([tokens_features,trigrams_features, 
                         frames_features, entities_features, 
                         sentiment_features, topic_features])
```

```python
labels = merged_df_cleaned['nlp_label_Likely to be Bias']
```

```python
from sklearn.feature_selection import SelectKBest, chi2

k = 20000

# Apply SelectKBest using chi-squared test
selector = SelectKBest(score_func=chi2, k=k)

# Fit and transform your final_features
X_selected = selector.fit_transform(final_features, labels)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_selected, labels, test_size=0.2, random_state=42, stratify=labels
)

from sklearn.svm import LinearSVC

linear_svm = LinearSVC(class_weight='balanced', random_state=42, max_iter=10000)
linear_svm.fit(X_train, y_train)


from sklearn.metrics import classification_report, accuracy_score

y_pred = linear_svm.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
poly_svm = SVC(
    kernel='poly',
    class_weight='balanced',
    random_state=42
)
poly_svm.fit(X_train, y_train)
```

```python
y_pred = poly_svm.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
rbf_svm = SVC(
    kernel='rbf',
    class_weight='balanced',
    random_state=42
)
rbf_svm.fit(X_train, y_train)
```

```python
y_pred = rbf_svm.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
