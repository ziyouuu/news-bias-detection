```python
import pandas as pd
df = pd.read_csv('../datasets/news_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['image', 'nlp-image_label'])
df.head()
```

```python
#lowercase
df['article_text'] = df['article_text'].apply(lambda x: x.lower())
```

```python
# remove punctuation
import re

def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df['article_text'] = df['article_text'].apply(remove_punctuation)
```

```python
#encoding for label 

df = pd.get_dummies(df, columns=['nlp_label'], dtype=int)
df.head()
```

```python
#tokenise
import nltk
from nltk.tokenize import word_tokenize

df['article_text_tokens'] = df['article_text'].apply(word_tokenize)
```

```python
# stopwords
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
article_text_tokens_no_stopwords = []
for row in df['article_text_tokens']:
    filtered_row = []
    for w in row:
        if w not in stop_words:
            filtered_row.append(w)
    article_text_tokens_no_stopwords.append(filtered_row)

df['article_text_tokens_no_stopwords'] = article_text_tokens_no_stopwords
```

```python
# bigram trigram 
from nltk.util import ngrams

def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))

# Generate n-grams for n = 2 (bigrams) and n = 3 (trigrams)
df['article_text_bigrams'] = df['article_text_tokens_no_stopwords'].apply(lambda x: generate_ngrams(x, 2))
df['article_text_trigrams'] = df['article_text_tokens_no_stopwords'].apply(lambda x: generate_ngrams(x, 3))
```

```python
df.head()
```

```python
sentiments_df = pd.read_csv('../datasets/generated/news_dataset_sentiment_labels.csv')
```

```python
# Merge the DataFrames on the 'unique_id' column
merged_df = pd.merge(df, sentiments_df[['unique_id', 'negative', 'neutral', 'positive']], on='unique_id', how='left')
```

```python
merged_df.info()
```

```python
merged_df_cleaned = merged_df.dropna()
```

```python
import pandas as pd
merged_df_cleaned = pd.read_csv('../datasets/generated/bias_detection_news_dataset_content_cleaned.csv')
merged_df_cleaned.head()
```

```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.experimental import enable_halving_search_cv 
from sklearn.model_selection import HalvingGridSearchCV
model_map = {"svm_linear": SVC(kernel = "linear"),
             "svm_rbf": SVC(kernel = "rbf"),
            "svm_poly": SVC(kernel = "poly")}

parameters_map = { 
    "C": [0.01, 0.1, 1, 10, 100],
    "gamma": [0.001, 0.01, 0.1, 1],
    "degree": [2, 3, 4, 5] 
}

def fine_tune(model: str, param_grid: dict, folds:int = 3):
    return HalvingGridSearchCV(model_map[model], param_grid, n_jobs = -1, refit = True, cv = folds, verbose = 3, factor = 3, resource = "n_samples")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Function to join lists into strings for vectorization
def join_lists_to_string(list_column):
    return [' '.join(['_'.join(tup) if isinstance(tup, tuple) else str(tup) for tup in lst]) for lst in list_column]

# Convert list columns to strings
merged_df_cleaned['article_text_tokens_no_stopwords_str'] = join_lists_to_string(merged_df_cleaned['article_text_tokens_no_stopwords'])
merged_df_cleaned['article_text_bigrams_str'] = join_lists_to_string(merged_df_cleaned['article_text_bigrams'])
merged_df_cleaned['article_text_trigrams_str'] = join_lists_to_string(merged_df_cleaned['article_text_trigrams'])

# Initialize vectorizers
vectorizer_tokens = TfidfVectorizer()
vectorizer_bigrams = TfidfVectorizer()
vectorizer_trigrams = TfidfVectorizer()

# Fit and transform the data
tokens_features = vectorizer_tokens.fit_transform(merged_df_cleaned['article_text_tokens_no_stopwords_str'])
bigrams_features = vectorizer_bigrams.fit_transform(merged_df_cleaned['article_text_bigrams_str'])
trigrams_features = vectorizer_trigrams.fit_transform(merged_df_cleaned['article_text_trigrams_str'])

# Combine as sparse matrix (no conversion to dense)
from scipy.sparse import hstack

combined_features = hstack([tokens_features, bigrams_features])

# Convert sentiment features to sparse matrix before concatenating
from scipy.sparse import csr_matrix

sentiment_features = csr_matrix(merged_df_cleaned[['negative', 'neutral', 'positive']].values)

# Final concatenation in sparse format
final_features = hstack([combined_features, sentiment_features])
```

```python
labels = merged_df_cleaned['nlp_label_Likely to be Bias']
```

```python
train_paths, test_paths, train_labels, test_labels = train_test_split(
    final_features, labels, test_size=0.2, random_state=42, stratify=labels
)
```

```python
linear_svm = fine_tune("svm_linear", parameters_map)
linear_svm.fit(train_paths[:1000], train_labels[:1000])
```

```python
y_pred = linear_svm.predict(test_paths[:1000])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

def prediction_score(test_labels, y_pred):
    # Calculate accuracy
    accuracy = accuracy_score(test_labels, y_pred)  
    print(f"Accuracy: {accuracy * 100:.2f}%")

    # Calculate confusion matrix
    conf_matrix = confusion_matrix(test_labels, y_pred)
    print("Confusion Matrix:")
    print(conf_matrix)

    # Calculate precision, recall, and F1 score
    precision = precision_score(test_labels, y_pred)
    recall = recall_score(test_labels, y_pred)
    f1 = f1_score(test_labels, y_pred)

    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")

prediction_score(test_labels[:1000], y_pred)
```

```python
poly_svm = fine_tune("svm_poly", parameters_map)
poly_svm.fit(train_paths[:1000], train_labels[:1000])
```

```python
print(f"Grid search poly SVM best parameters: {poly_svm.best_params_}")
print(f"Cross validated score of the best poly SVM model: {poly_svm.best_score_}")
```

```python
y_pred_poly = poly_svm.predict(test_paths[:1000])
```

```python
prediction_score(test_labels[:1000], y_pred_poly)
```

```python
# rbf
rbf_svm = fine_tune("svm_rbf", parameters_map)
rbf_svm.fit(train_paths[:1000], train_labels[:1000])
```

```python
# Check the best parameters
print("Best parameters found:", rbf_svm.best_params_)

# Check the kernel of the best estimator
print("Kernel used:", rbf_svm.best_estimator_.kernel)
```

```python
y_pred_rbf = rbf_svm.predict(test_paths[:1000])
```

```python
prediction_score(test_labels[:1000], y_pred_rbf)
```
