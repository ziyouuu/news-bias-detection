```python
import pandas as pd
df = pd.read_csv('../datasets/news_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['image', 'nlp-image_label'])
df.head()
```

```python
#lowercase
df['article_text'] = df['article_text'].apply(lambda x: x.lower())
```

```python
# remove punctuation
import re

def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df['article_text'] = df['article_text'].apply(remove_punctuation)
```

```python
#encoding for label 

df = pd.get_dummies(df, columns=['nlp_label'], dtype=int)
df.head()
```

```python
#tokenise
import nltk
from nltk.tokenize import word_tokenize

df['article_text_tokens'] = df['article_text'].apply(word_tokenize)
```

```python
# stopwords
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
article_text_tokens_no_stopwords = []
for row in df['article_text_tokens']:
    filtered_row = []
    for w in row:
        if w not in stop_words:
            filtered_row.append(w)
    article_text_tokens_no_stopwords.append(filtered_row)

df['article_text_tokens_no_stopwords'] = article_text_tokens_no_stopwords
```

```python
sentiments_df = pd.read_csv('news-bias-detection/datasets/generated/news_dataset_sentiment_labels.csv')
# "C:\Users\ZY\Documents\Y3S2\IS450 NLP\Project\news-bias-detection\datasets\generated\news_dataset_sentiment_labels.csv"
```

```python
# Merge the DataFrames on the 'unique_id' column
merged_df = pd.merge(df, sentiments_df[['unique_id', 'negative', 'neutral', 'positive']], on='unique_id', how='left')
```

```python
merged_df_cleaned = merged_df.dropna()
```

```python
#Create dictionary
import gensim
from gensim import corpora
all_articles = merged_df_cleaned['article_text_tokens_no_stopwords'].tolist()  # Convert to a list of lists
dictionary = corpora.Dictionary(all_articles)
print(dictionary)
```

```python
# Convert all documents to TF Vectors
all_tf_vectors = [dictionary.doc2bow(article) for article in all_articles]
```

```python
# Step 3: Convert TF vectors to dictionary representation
all_data_as_dict = [{id: 1 for (id, tf_value) in vec} for vec in all_tf_vectors]

# # Step 4: Get labels from the DataFrame
# labels = merged_df_cleaned['nlp_label_Likely to be Bias'].tolist()  # Adjust the column name if needed

# # Step 5: Combine data with labels
# all_data_with_labels = [(data_dict, label) for data_dict, label in zip(all_data_as_dict, labels)]
```

```python
# Extract sentiment scores
sentiment_scores = merged_df_cleaned[['negative', 'neutral', 'positive']].values.tolist()

# Combine TF features with sentiment scores
all_data_with_sentiment = []
for i in range(len(all_data_as_dict)):
    feature_dict = all_data_as_dict[i]  # Get word-based features
    feature_dict['negative'] = sentiment_scores[i][0]  
    feature_dict['neutral'] = sentiment_scores[i][1]
    feature_dict['positive'] = sentiment_scores[i][2]
    all_data_with_sentiment.append(feature_dict)
```

```python
import nltk

# Get labels
labels = merged_df_cleaned['nlp_label_Likely to be Bias'].tolist()

# Combine data with labels
all_data_with_labels = [(data_dict, label) for data_dict, label in zip(all_data_with_sentiment, labels)]

# Train Naïve Bayes Classifier
classifier = nltk.NaiveBayesClassifier.train(all_data_with_labels)

# Evaluate Accuracy
accuracy = nltk.classify.accuracy(classifier, all_data_with_labels)
print("Accuracy:", accuracy)
```
