```python
import pandas as pd
df = pd.read_csv('../datasets/news_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['image', 'nlp-image_label'])
df.head()
```

```python
#lowercase
df['article_text'] = df['article_text'].apply(lambda x: x.lower())
```

```python
# remove punctuation
import re

def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df['article_text'] = df['article_text'].apply(remove_punctuation)
```

```python
#encoding for label 

df = pd.get_dummies(df, columns=['nlp_label'], dtype=int)
df.head()
```

```python
#tokenise
import nltk
from nltk.tokenize import word_tokenize

df['article_text_tokens'] = df['article_text'].apply(word_tokenize)
```

```python
# stopwords
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
article_text_tokens_no_stopwords = []
for row in df['article_text_tokens']:
    filtered_row = []
    for w in row:
        if w not in stop_words:
            filtered_row.append(w)
    article_text_tokens_no_stopwords.append(filtered_row)

df['article_text_tokens_no_stopwords'] = article_text_tokens_no_stopwords
```

```python
#Create dictionary
import gensim
from gensim import corpora
all_articles = merged_df_cleaned['article_text_tokens_no_stopwords'].tolist()  # Convert to a list of lists
dictionary = corpora.Dictionary(all_articles)
print(dictionary)
```

```python
# Convert all documents to TF Vectors
all_tf_vectors = [dictionary.doc2bow(article) for article in all_articles]
```

```python
# Step 3: Convert TF vectors to dictionary representation
all_data_as_dict = [{id: 1 for (id, tf_value) in vec} for vec in all_tf_vectors]
```

```python
import nltk
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

labels = merged_df_cleaned['nlp_label_Likely to be Bias'].tolist()

all_data_with_labels = [(data_dict, label) for data_dict, label in zip(all_data_as_dict, labels)]

train_data, test_data = train_test_split(all_data_with_labels, test_size=0.2, random_state=42)

classifier = nltk.NaiveBayesClassifier.train(train_data)

test_features = [features for features, label in test_data]
test_labels = [label for features, label in test_data]

predicted_labels = [classifier.classify(features) for features in test_features]

accuracy = nltk.classify.accuracy(classifier, test_data)
print("Accuracy on Test Data:", accuracy)

conf_matrix = confusion_matrix(test_labels, predicted_labels)
print("Confusion Matrix:\n", conf_matrix)

class_report = classification_report(test_labels, predicted_labels)
print("Classification Report:\n", class_report)
```
